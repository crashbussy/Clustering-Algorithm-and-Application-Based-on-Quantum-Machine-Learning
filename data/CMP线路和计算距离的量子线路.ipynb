```python
#CMP线路
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, transpile 
from qiskit.visualization import circuit_drawer  
a = QuantumRegister(1, name='a') 
b = QuantumRegister(1, name='b') 
c = QuantumRegister(1, name='c') 
o = QuantumRegister(2, name='output')  #o = output 
# Create the quantum circuit 
circuit = QuantumCircuit(a, b,c, o)  
# Comparison circuit 
circuit.x(b[0])
circuit.x(o[1])
circuit.ccx(a[0],b[0],c[0])  
circuit.ccx(o[1],c[0],o[0])
circuit.ccx(a[0],b[0],c[0])  
circuit.cx(o[0],o[1])
circuit.x(b[0]) 
circuit.x(o[1])
circuit.ccx(a[0],b[0],c[0])  
circuit.ccx(o[1],c[0],o[0])
circuit.ccx(a[0],b[0],c[0]) 
#Draw the circuit 
circuit_transpiled = transpile(circuit, basis_gates=['cx', 'u3']) 
circuit_drawer(circuit_transpiled, output='mpl')
#circuit_drawer(circuit, output='mpl')
```

```python
#(计算距离用的线路)
from math import pi
theta_1 = 0
theta_2 = 0

qr = QuantumRegister(3, name="qr")

cr = ClassicalRegister(1, name="cr")

qc = QuantumCircuit(qr, cr, name="k_means")   
qc.h(qr[0])
qc.h(qr[1])
qc.h(qr[2])
qc.u(theta_1, pi, pi, qr[1])
qc.u(theta_2, pi, pi, qr[2])
qc.cswap(qr[0], qr[1], qr[2])
qc.h(qr[0])
qc.measure(qr[0], cr[0])
qc.reset(qr)
circuit_transpiled = transpile(qc) 
circuit_drawer(circuit_transpiled,output='mpl')
```
